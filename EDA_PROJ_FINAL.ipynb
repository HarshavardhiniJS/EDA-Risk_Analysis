### Converting negative age to positive

```python
tmp_column_names = [ 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'DAYS_REGISTRATION', 'DAYS_ID_PUBLISH', 'DAYS_LAST_PHONE_CHANGE']
for tmp_col in tmp_column_names:
  print('\n---')
  print("Column : ", tmp_col)
  print(df[tmp_col].head())
  print("Converting negative age to positive using abs()")
  df[tmp_col]=abs(df[tmp_col])
  print(df[tmp_col].head())
```

# Handling outliers


The numerical columns in the dataframe are to be listed and analysed for checking outliers

```python
num_cols=["AMT_INCOME_TOTAL","AMT_CREDIT","AMT_ANNUITY","AMT_GOODS_PRICE","DAYS_EMPLOYED","DAYS_REGISTRATION","CNT_FAM_MEMBERS","DAYS_LAST_PHONE_CHANGE"]
```

```python
for cols in num_cols:
    q1=df[cols].describe()["25%"]
    q3=df[cols].describe()["75%"]
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    print("\n---")
    print("column {} : {}  {} ".format(cols,lower_bound,upper_bound))
    val_out_bounds_tmp= df[(df[cols]<lower_bound)|(df[cols]>upper_bound)][cols].count()
    tot_bound_tmp = df[cols].count()
    perce_tmp = (val_out_bounds_tmp/tot_bound_tmp)*100
    print("Count of values outside bounds: {} = {}%".format(val_out_bounds_tmp, round(perce_tmp,2)))
    print("Total values in column: {}".format(tot_bound_tmp))

```

# Handling outliers

The numerical columns in the dataframe are to be listed and analysed for checking outliers

```python
p_num_cols=["AMT_ANNUITY","AMT_CREDIT","AMT_GOODS_PRICE","AMT_APPLICATION"]
```

```python
for cols in p_num_cols:
    q1=prev_df[cols].describe()["25%"]
    q3=prev_df[cols].describe()["75%"]
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    print("\n---")
    print("column {} : {}  {} ".format(cols,lower_bound,upper_bound))
    val_out_bounds_tmp= prev_df[(prev_df[cols]<lower_bound)|(prev_df[cols]>upper_bound)][cols].count()
    tot_bound_tmp = prev_df[cols].count()
    perce_tmp = (val_out_bounds_tmp/tot_bound_tmp)*100
    print("Count of values outside bounds: {} = {}%".format(val_out_bounds_tmp, round(perce_tmp,2)))
    print("Total values in column: {}".format(tot_bound_tmp))
```
